```python
import random

def division():
    # Guardar valores aleatorios en variables
    random_1 = random.randint(1, 100)
    random_2 = random.randint(1, 100)
    random_3 = random.randint(1, 100) 
    
    # Guardar en una lista los valores aleatorios
    lista = [random_1, random_2, random_3]
    
    # Random divisor
    divisor = random.randint(1, 10)
    
    # Crear lista del divisor 
    lista_divisor = [divisor, divisor, divisor]

    # Cocientes y su lista
    cociente_1 = random_1 // divisor
    cociente_2 = random_2 // divisor
    cociente_3 = random_3 // divisor
    lista_cociente = [cociente_1, cociente_2, cociente_3]

    # Residuos y su lista
    residuo_1 = random_1 % divisor
    residuo_2 = random_2 % divisor
    residuo_3 = random_3 % divisor
    lista_residuo = [residuo_1, residuo_2, residuo_3]
    
    # Crear diccionario
    diccionario_division = {'Dividendo': lista, 'Divisor': lista_divisor, 'Cociente': lista_cociente, 'Residuo': lista_residuo}
    return diccionario_division

#Llamar a la funcion
print(division())
```

```python
#Crear funcion
def contar_palabras(frase):
    #Crear una lista con las paralbras de la frase
    lista_palabras = frase.split()

    #Crear diccionario return 
    diccionario_repetidas = {lista_palabras[0]: 1, lista_palabras[1]: 1, lista_palabras[2]: 1, lista_palabras[3]: 1, lista_palabras[4]: 1 }
    
    return diccionario_repetidas

print(contar_palabras("¡Hola!, ¿cómo va todo? Éxitos"))
```

# Ejercicio 4

```python
def caracter_presente(frase, caracter):
    # Hacer la frase en una lista de 2 strings (palabras)
    lista_str = frase.split()

    # Caso cuando ambas letras tiene el caracter
    if lista_str[0].find(caracter) >= 0 and lista_str[1].find(caracter) >= 0:
        if len(lista_str[0]) > len(lista_str[1]):
            retornar = "'" + caracter + "': " + lista_str[0]
            return retornar
        elif len(lista_str[0]) < len(lista_str[1]):
            retornar = "'" + caracter + "': " + lista_str[1]
            return retornar
        #si ambas tiene letras son del mismo tamaño
        else:
            retornar = "'" + caracter + "': " + lista_str[0] + ", " + lista_str[1]
            return retornar
            
    #Caso cuando solo una de las palabras tiene el caracter      
    elif lista_str[0].find(caracter) < 0 and lista_str[1].find(caracter) >= 0:
        retornar = "'" + caracter + "': " + lista_str[1]
        return retornar
    else:
        retornar = "'" + caracter + "': " + lista_str[0]
        return retornar
        
print(caracter_presente("Ok amigos", "i"))
print(caracter_presente("Hola mundo", "o"))
```

```python
import datetime

lista_pacientes = {
    "1": {
        "Edad": 23,
        "Tipo sangre": "-O",
        "Fecha ingreso": datetime.datetime(2023, 6, 1),
        "Fecha salida": datetime.datetime(2023, 6, 15),
        "Nivel de emergencia": 2,
        "Pago de servicios": 502333
    },
    "2": {
        "Edad": 7,
        "Tipo sangre": "AB+",
        "Fecha ingreso": datetime.datetime(2024, 1, 1),
        "Fecha salida": None,
        "Nivel de emergencia": 5,
        "Pago de servicios": 0
    },
    "3": {
        "Edad": 65,
        "Tipo sangre": "AB+",
        "Fecha ingreso": datetime.datetime(2023, 2, 1),
        "Fecha salida": datetime.datetime(2024, 2, 1),
        "Nivel de emergencia": 4,
        "Pago de servicios": 28006000
    },
    "4": {
        "Edad": 23,
        "Tipo sangre": "AB-",
        "Fecha ingreso": datetime.datetime(2024, 2, 1),
        "Fecha salida": datetime.datetime(2024, 2, 1),
        "Nivel de emergencia": 0,
        "Pago de servicios": 200300
    },
}

def caracteristicas_pacientes(lista_dicc):
    # Calcular edad promedio
    edad_promedio = (lista_dicc["1"]["Edad"] + lista_dicc["2"]["Edad"] + lista_dicc["3"]["Edad"] + lista_dicc["4"]["Edad"]) / 4

    # Calcular tipo de sangre más común
    contador_O = contador_AB_positivo = contador_AB_negativo = 0
    if lista_dicc["1"]["Tipo sangre"] == "-O":
        contador_O += 1
    if lista_dicc["2"]["Tipo sangre"] == "-O":
        contador_O += 1
    if lista_dicc["3"]["Tipo sangre"] == "-O":
        contador_O += 1
    if lista_dicc["4"]["Tipo sangre"] == "-O":
        contador_O += 1
    if lista_dicc["1"]["Tipo sangre"] == "AB+":
        contador_AB_positivo += 1
    if lista_dicc["2"]["Tipo sangre"] == "AB+":
        contador_AB_positivo += 1
    if lista_dicc["3"]["Tipo sangre"] == "AB+":
        contador_AB_positivo += 1
    if lista_dicc["4"]["Tipo sangre"] == "AB+":
        contador_AB_positivo += 1
    if lista_dicc["1"]["Tipo sangre"] == "AB-":
        contador_AB_negativo += 1
    if lista_dicc["2"]["Tipo sangre"] == "AB-":
        contador_AB_negativo += 1
    if lista_dicc["3"]["Tipo sangre"] == "AB-":
        contador_AB_negativo += 1
    if lista_dicc["4"]["Tipo sangre"] == "AB-":
        contador_AB_negativo += 1

    if contador_O > contador_AB_positivo and contador_O > contador_AB_negativo:
        sangre_comun = "-O"
    elif contador_AB_positivo > contador_O and contador_AB_positivo > contador_AB_negativo:
        sangre_comun = "AB+"
    else:
        sangre_comun = "AB-"

    # Calcular pago promedio
    pago_promedio = (lista_dicc["1"]["Pago de servicios"] + lista_dicc["2"]["Pago de servicios"] + lista_dicc["3"]["Pago de servicios"] + lista_dicc["4"]["Pago de servicios"]) / 4

    # Calcular tiempo en el hospital de cada paciente
    if lista_dicc["1"]["Fecha salida"] is not None:
        tiempo_1 = lista_dicc["1"]["Fecha salida"] - lista_dicc["1"]["Fecha ingreso"]
    else:
        tiempo_1 = "NA"
    
    if lista_dicc["2"]["Fecha salida"] is not None:
        tiempo_2 = lista_dicc["2"]["Fecha salida"] - lista_dicc["2"]["Fecha ingreso"]
    else:
        tiempo_2 = "NA"
    
    if lista_dicc["3"]["Fecha salida"] is not None:
        tiempo_3 = lista_dicc["3"]["Fecha salida"] - lista_dicc["3"]["Fecha ingreso"]
    else:
        tiempo_3 = "NA"
    
    if lista_dicc["4"]["Fecha salida"] is not None:
        tiempo_4 = lista_dicc["4"]["Fecha salida"] - lista_dicc["4"]["Fecha ingreso"]
    else:
        tiempo_4 = "NA"


    # Calcular pago según nivel de emergencia
    emergencia_monto = {
        lista_dicc["1"]["Nivel de emergencia"]: lista_dicc["1"]["Pago de servicios"],
        lista_dicc["2"]["Nivel de emergencia"]: lista_dicc["2"]["Pago de servicios"],
        lista_dicc["3"]["Nivel de emergencia"]: lista_dicc["3"]["Pago de servicios"],
        lista_dicc["4"]["Nivel de emergencia"]: lista_dicc["4"]["Pago de servicios"]
    }

    return [
        "Edad promedio: " + str(edad_promedio),
        "Sangre más común: " + sangre_comun,
        "Pago promedio: " + str(pago_promedio),
        "Tiempo en hospital según el paciente: Paciente 1: " + str(tiempo_1) + ", Paciente 2: " + str(tiempo_2) + ", Paciente 3: " + str(tiempo_3) + ", Paciente 4: " + str(tiempo_4),
        "Pago según nivel de emergencia: " + str(emergencia_monto)
    ]

print(caracteristicas_pacientes(lista_pacientes))
```

# Ejercicio 6

```python
import datetime
#Crear una funcion que determine el tiempo de cada cliente
def tiempo(lista_dicc):
    #Identificar el tiempo en el hospital
    if lista_dicc["1"]["Fecha salida"] is not None:
        tiempo_1 = lista_dicc["1"]["Fecha salida"] - lista_dicc["1"]["Fecha ingreso"]
    else:
        tiempo_1 = datetime.datetime(2024, 3, 28) - lista_dicc["1"]["Fecha ingreso"]  #si el tiempo de salida es NA se toma la fecha de hoy 
     
    if lista_dicc["2"]["Fecha salida"] is not None:
        tiempo_2 = lista_dicc["2"]["Fecha salida"] - lista_dicc["2"]["Fecha ingreso"]
    else:
        tiempo_2 = datetime.datetime(2024, 3, 28) - lista_dicc["2"]["Fecha ingreso"] #si el tiempo de salida es NA se toma la fecha de hoy 
         
    if lista_dicc["3"]["Fecha salida"] is not None:
        tiempo_3 = lista_dicc["3"]["Fecha salida"] - lista_dicc["3"]["Fecha ingreso"]
    else:
        tiempo_3 = datetime.datetime(2024, 3, 28) - lista_dicc["3"]["Fecha ingreso"] #si el tiempo de salida es NA se toma la fecha de hoy 
        
    if lista_dicc["4"]["Fecha salida"] is not None:
        tiempo_4 = lista_dicc["4"]["Fecha salida"] - lista_dicc["4"]["Fecha ingreso"]
    else:
        tiempo_4 = datetime.datetime(2024, 3, 28) - lista_dicc["4"]["Fecha ingreso"] #si el tiempo de salida es NA se toma la fecha de hoy 
        
    return [tiempo_1, tiempo_2, tiempo_3, tiempo_4]


#Crear funcion que tiene como entrada la edad, y el tiempo en el hospi, para devolver el costo 
def costo(edad, tiempo):
    if edad < 18:
        return 0
    else:
        if tiempo.days < 1:
            return 8000
        elif tiempo.days < 30:
            return tiempo.days * 15000
        else:
            meses_completos = tiempo.days // 30
            return meses_completos * 1000000


def cobro_gerencia(lista_dicc):
    lista_tiempos = tiempo(lista_dicc)
    #Determinar el costo de cada persona con la funcion creada anteriormente
    costo_1 = costo(lista_dicc["1"]["Edad"], lista_tiempos[0])
    costo_2 = costo(lista_dicc["2"]["Edad"], lista_tiempos[1])
    costo_3 = costo(lista_dicc["3"]["Edad"], lista_tiempos[2])
    costo_4 = costo(lista_dicc["4"]["Edad"], lista_tiempos[3])
    
    #Obeter diferencias respecto al pago acutal
    dife_1 = costo_1 * (1 + 0.13) - lista_dicc["1"]["Pago de servicios"]
    dife_2 = costo_2 * (1 + 0.13) - lista_dicc["2"]["Pago de servicios"]
    dife_3 = costo_3 * (1 + 0.13) - lista_dicc["3"]["Pago de servicios"] 
    dife_4 = costo_4 * (1 + 0.13) - lista_dicc["4"]["Pago de servicios"] 
    
    #Crear diccionarios por persona
    dicc_per1 = {"Pagaría": costo_1 * (1 + 0.13), "Diferencia respecto a pago actual": dife_1 , "IVA": costo_1 * 0.13}
    dicc_per2 = {"Pagaría": costo_2 * (1 + 0.13), "Diferencia respecto a pago actual": dife_2 , "IVA": costo_2 * 0.13}
    dicc_per3 = {"Pagaría": costo_3 * (1 + 0.13), "Diferencia respecto a pago actual": dife_3 , "IVA": costo_3 * 0.13}
    dicc_per4 = {"Pagaría": costo_4 * (1 + 0.13), "Diferencia respecto a pago actual": dife_4 , "IVA": costo_4 * 0.13}

    return ["Persona 1: " + str(dicc_per1),"Persona 2: " + str(dicc_per2), "Persona 3: " + str(dicc_per3), "Persona 4: " + str(dicc_per4)]

print(cobro_gerencia(lista_pacientes))

```

# Ejercicio 7

```python
def prima(tipo_seguro, valor, meses_cobertura):

     #inicializar el diccionario con las clases de seguros
    seguros = {"Automóviles comunes": 0.03, "Carros para Uber": 0.05, "Motocicletas": 0.055, "Camión de carga pesada o liviana": 0.04, "Buses": 0.05, "Embarcaciones": 0.05}

    #Verificar el tipo de seguro esta en el diccionario
    if tipo_seguro not in seguros:
        return "Tipo de seguro no existe"
    else:
        #Calcular la prima base, que es la prima por el valor fiscal
        pago_mensual = seguros[tipo_seguro] * valor
    
        

        #Crear diccionario donde se ingrese el plazo y devuelva una lista con la cantidad de meses de cada plzo y la prima de ese plazo
       # dicc_primas = {"mensual": [1, pago_mensual], "trimestral": [3,pago_mensual * 3], "semestral":[6, pago_mensual * 6], "anual": [12, pago_mensual * 12]}


        if meses_cobertura == 0:
            return "Meses de conbertura inválido"
        else:
            prima_descuento = pago_mensual * (1 - (meses_cobertura - 1) * 0.005) 
            monto_total = prima_descuento * meses_cobertura
            
        
        #Calcular la prima bajo el descuento, que es aplicarle descuento al pago mensual
        #con_descuento = dicc_primas[plazo][1] * (1 - 0.005 * adicional_meses)
    
            
        #Diccionario que devuelve lo solicitado    
        dicc_return = {'Tipo de seguro': tipo_seguro, 
                       'Prima base': pago_mensual, 
                       'Prima bajo descuento': prima_descuento, 
                       'Meses de cobertura': meses_cobertura,
                       'Pago total': monto_total
                      }
        return dicc_return

print(prima("Motocicletas", 2000000 , 6))
```

# Ejercicio 8

```python
#Se va a tomar a garantia como un boolean pa
def credito_empresaril(monto_solicitado, plazo_anual, moneda, rating, garantia, valor_garantia):

    #Primero determinar la tasa de interés segun la calificacion
    #obtengamos el valor del z= (BBB-AAA)/3
    z = (0.084 - 0.064)/3 #entre AAA y BBB
    y = (0.15 - 0.084)/3 #entre C y BBB
    
    if rating == "AAA":
        tasa_base = 0.064
        
    elif rating == "BBB":
        tasa_base = 0.084
        
    elif rating == "C":
        tasa_base = 0.15
        
    elif rating == "D":
        tasa_base = 0.25
        
    elif rating == "AA":
        tasa_base = 0.064 + z
        
    elif rating == "A":
        tasa_base = 0.064 + 2*z
        
    elif rating == "BB": 
        tasa_base = 0.084 + y

    elif rating == "B":
        tasa_base = 0.084 + 2*y
        
    else:
        return "Calificación no valida"

    #Inicializar la variable tasa neta, que es la tasa final
    tasa_neta = tasa_base

    
    #Verificar si tiene garantía o no
    if garantia == True:
        tasa_preferencial = 0.01
        tasa_neta = tasa_neta - tasa_preferencial
    else:
        tasa_preferencial = 0

    #Verificar si es en moneda extranjera y asígnar tasa
    if moneda == "colones":
        tasa_moneda_ext = 0
    else:
        tasa_moneda_ext = 0.035
        tasa_neta = tasa_neta + tasa_moneda_ext

    #Determinar comisiones
    comisiones = monto_solicitado * 0.05

    #Monto del prestamo total es lo solicitado mas comisiones
    monto_total = monto_solicitado + comisiones 

    #Segun google el LTV se calcula así
    LTV = (monto_total / valor_garantia) * 100 

    plazo_meses = plazo_anual * 12

    #Se tiene que la tasa neta es anual y ademas debe amortizar el préstamo, entonces es la suma de la tasa mensual mas amortizacion 
    
    tasa_interes_mensual =  (( 1  +  tasa_neta ) ** ( 1  /  12 )  -  1 )    #suponiendo que es la tasa anual dada es efectiva y se pasa a efectiva mensual

    cuota_mensual  =  monto_total  *  tasa_interes_mensual /  ( 1  -  ( 1  +  tasa_interes_mensual )  **  - plazo_meses ) # Formula L=Ra

   

    #diccionario a retornar
    dicc_return ={
        'Empresa' : "NA",
        'Rating' : rating,
        'Tasa de interés base' : tasa_base,
        'Moneda del préstamo' : moneda,
        'Tasa preferencial' : tasa_preferencial,
        'Tasa por moneda extranjera' : tasa_moneda_ext,
        'Tasa neta' : tasa_neta,
        'Monto préstamo solicitado' : monto_solicitado,
        'Comisiones' : comisiones,
        'Monto del préstamo total' : monto_total,
        'LTV' : str(LTV) + "%",
        'Cantidad de meses a pagar' : plazo_meses,
        'Cuota mensual' : cuota_mensual       
    }

    return dicc_return

print(credito_empresaril(1200000, 30, "dolares", "A", True , 1500000))
    
        
    
```
